```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("telcochurnda.csv")
```

```python
df.shape
df.info()
df.describe(include="all")
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df[df["TotalCharges"].isnull()][["tenure", "TotalCharges"]].head()
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df["ChurnFlag"] = df["Churn"].map({"YES":1,"NO":0})
```

```python
df["ChurnFlag"].value_counts(normalize=True)
```

```python
df["Churn"].value_counts()
```

```python
df["ChurnFlag"].value_counts(dropna=False)
```

```python
df["Churn"] = df["Churn"].astype(str).str.strip()
df["ChurnFlag"] = df["Churn"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df["ChurnFlag"].value_counts()
```

```python
df.groupby("Contract")["ChurnFlag"].mean().sort_values(ascending=False)
```

```python
bins = [0, 6, 12, 24, 100]
labels = ["0-6 months", "6-12 months", "12-24 months", "24+ months"]

df["TenureBucket"] = pd.cut(df["tenure"], bins=bins, labels=labels, right=False)
```

```python
df.groupby("TenureBucket")["ChurnFlag"].mean()
```
